```python
# Dependencies
import pandas as pd
#import os
#import sys
import csv
import json
```

```python
# Load in csv file
dallas_open_data = 'dataset_dallas_open_data/Restaurant_and_Food_Establishment_Inspections__October_2016_to_Present_.csv'
```

```python
# Read and display the CSV with Pandas
dallas_open_data_pd = pd.read_csv(dallas_open_data)
#dallas_open_data_pd.head()
```

```python
# Extract specific columns for analysis

restaurant_name = dallas_open_data_pd['Restaurant Name'].str.strip()
inspection_type = dallas_open_data_pd['Inspection Type'].str.strip()
inspection_date = dallas_open_data_pd['Inspection Date'].str.strip()
inspection_score = dallas_open_data_pd['Inspection Score']
street_number = dallas_open_data_pd['Street Number'].astype('str').str.strip()
street_name = dallas_open_data_pd['Street Name'].str.strip()
street_unit = dallas_open_data_pd['Street Unit'].fillna('').astype('str')
zip_code = dallas_open_data_pd['Zip Code']
street_address = dallas_open_data_pd['Street Address'].str.strip()
inspection_month = dallas_open_data_pd['Inspection Month'].str.strip()
inspection_year = dallas_open_data_pd['Inspection Year'].str.strip()

# Check record count.  Should be 42,070
len(restaurant_name)
```

```python
#Save data tp dataframe
clean_dallas_open_data_pd = pd.DataFrame({
    'restaurant_name':restaurant_name,
    'inspection_type':inspection_type,
    'inspection_date':inspection_date,
    'inspection_score':inspection_score,
    'street_number':street_number,
    'street_name':street_name,
    'street_unit':street_unit,
    'zip_code':zip_code,
    'street_address':street_address,
    'inspection_month':inspection_month,
    'inspection_year':inspection_year
})

# Display dataframe
clean_dallas_open_data_pd
```
