```python
counter = 0

# yelp_response = []

# for restaurant in dallas_open_data_df_limited:
for row in dallas_open_data_df_limited["restaurant_name"]:
    
    # get restaurant data from df and set as variable to pass through to API call
    # restaurant search variables
    search_name = dallas_open_data_df_limited["restaurant_name"][counter]
#     search_address = dallas_open_data_df_limited["street_address"][counter]
    search_street_number = dallas_open_data_df_limited["street_number"][counter]
    search_street_name = dallas_open_data_df_limited["street_name"][counter]
    search_zip_code = dallas_open_data_df_limited["zip_code"][counter]
    search_lat = dallas_open_data_df_limited["lat"][counter]
    search_lng = dallas_open_data_df_limited["long"][counter]
    
    # set parameters
    params = {"term": f"restaurant,{search_name}",
              "location": f"{search_street_number} {search_street_name}, {search_zip_code}",
#               "latitude": xx,
#               "longitude": xx,
              "radius": 100,
              "limit": 5
             }
    
    yelp_response = requests.get(url, params=params, headers=headers_1).json()
#     yelp_results = yelp_response["businesses"][0]
    yelp_results = yelp_response["businesses"]
    
    try:
        restaurant_category_test.append(yelp_results[0]["categories"][0]["title"])
    # not all results have > 1 category; in order to create dict, all lists must be same size - must assign "" when there is only one category
        if len(yelp_results[0]["categories"]) >= 1:
            restaurant_category_1.append(yelp_results[0]["categories"][0]["title"])
        else:
            restaurant_category_1.append(missing_api_result)
        if len(yelp_results[0]["categories"]) >= 2:
            restaurant_category_2.append(yelp_results[0]["categories"][1]["title"])
        else:
            restaurant_category_2.append(missing_api_result)
        if len(yelp_results[0]["categories"]) >= 3:
            restaurant_category_3.append(yelp_results[0]["categories"][2]["title"])
        else:
            restaurant_category_3.append(missing_api_result)
    except:
        restaurant_category_test.append(missing_api_result)
        
    try:
        restaurant_latitude.append(yelp_results[0]["coordinates"]["latitude"])
    except:
        restaurant_latitude.append(missing_api_result)
        
    try:
        restaurant_longitude.append(yelp_results[0]["coordinates"]["longitude"])
    except:
        restaurant_longitude.append(missing_api_result)
        
    try:
        restaurant_display_phone.append(yelp_results[0]["display_phone"])
    except:
        restaurant_display_phone.append(missing_api_result)
        
    try:
        restaurant_distance.append(yelp_results[0]["distance"])
    except:
        restaurant_distance.append(missing_api_result)
        
    try:
        restaurant_id.append(yelp_results[0]["id"])
    except:
        restaurant_id.append(missing_api_result)
        
    try:
        restaurant_alias.append(yelp_results[0]["alias"])
    except:
        restaurant_alias.append(missing_api_result)
        
    try:
        restaurant_image_url.append(yelp_results[0]["image_url"])
    except:
        restaurant_image_url.append(missing_api_result)
        
    try:
        restaurant_is_closed.append(yelp_results[0]["is_closed"])
    except:
        restaurant_is_closed.append(missing_api_result)
        
    try:
        restaurant_address1.append(yelp_results[0]["location"]["address1"])
    except:
        restaurant_address1.append(missing_api_result)
        
    try:
        restaurant_address2.append(yelp_results[0]["location"]["address2"])
    except:
        restaurant_address2.append(missing_api_result)
        
    try:
        restaurant_address3.append(yelp_results[0]["location"]["address3"])
    except:
        restaurant_address3.append(missing_api_result)
        
    try:
        restaurant_city.append(yelp_results[0]["location"]["city"])
    except:
        restaurant_city.append(missing_api_result)
        
    try:
        restaurant_country.append(yelp_results[0]["location"]["country"])
    except:
        restaurant_country.append(missing_api_result)
        
    try:
        restaurant_display_address.append(yelp_results[0]["location"]["display_address"])
    except:
        restaurant_display_address.append(missing_api_result)
        
    try:
        restaurant_state.append(yelp_results[0]["location"]["state"])
    except:
        restaurant_state.append(missing_api_result)
        
    try:
        restaurant_zip_code.append(yelp_results[0]["location"]["zip_code"])
    except:
        restaurant_zip_code.append(missing_api_result)
        
    try:
        restaurant_name.append(yelp_results[0]["name"])
    except:
        restaurant_name.append(missing_api_result)
        
    try:
        restaurant_phone.append(yelp_results[0]["phone"])
    except:
        restaurant_phone.append(missing_api_result)
        
#     restaurant_price.append(yelp_results[0]["price"])
    # not all results have Price; to prevent error, try/except error handling, assigning "", all lists must be same size - must assign "" when there is no Price
    try:
        restaurant_price.append(yelp_results[0]["price"])
    except:
        restaurant_price.append(missing_api_result)
    try:
        restaurant_rating.append(yelp_results[0]["rating"])
    except:
        restaurant_rating.append(missing_api_result)
    try:
        restaurant_review_count.append(yelp_results[0]["review_count"])
    except:
        restaurant_review_count.append(missing_api_result)
    try:
        restaurant_url.append(yelp_results[0]["url"])
    except:
        restaurant_url.append(missing_api_result)
    
    counter += 1
    
    
```
