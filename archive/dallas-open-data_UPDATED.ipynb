```python
# Dependencies
import pandas as pd
#import os
#import sys
import csv
import json
```

```python
# Load in csv file  
dallas_open_data = 'dallas_open_data/Restaurant_and_Food_Establishment_Inspections__October_2016_to_Present_.csv'
```

```python
# Read and display the CSV with Pandas
dallas_open_data_pd = pd.read_csv(dallas_open_data, low_memory=False)
#dallas_open_data_pd.head()
```

```python
# Extract specific columns for analysis
                                                             #(?:['-][a-zA-Z]+)
restaurant_name0 = dallas_open_data_pd['Restaurant Name']#.str.extract('([a-zA-Z0-9&+."-" ]+)', expand=False).str.strip()
#restaurant_name1 = dallas_open_data_pd['Restaurant Name'].str.extract('([a-zA-Z ]+)', expand=False).str.strip()

#best clean
restaurant_name2 = dallas_open_data_pd['Restaurant Name'].str.extract('([a-zA-Z0-9&+."\'""\-" ]+)', expand=False).str.strip()

restaurant_name3 = dallas_open_data_pd['Restaurant Name'].str.extract('([a-zA-Z0-9&+."\'""\-" ]+)', expand=False).str.strip()

#restaurant_name4 = dallas_open_data_pd['Restaurant Name'].str.replace(r'\([#)]*\)', '').str.strip()
inspection_type = dallas_open_data_pd['Inspection Type']

#inspection_date = dallas_open_data_pd['Inspection Date'].str.split()[0]

inspection_date = dallas_open_data_pd['Inspection Date'].astype('datetime64[ns]').dt.strftime('%m/%y')


inspection_score = dallas_open_data_pd['Inspection Score']
#max_inspection_score = dallas_open_data_pd.groupby(['Restaurant Name'])['Inspection Score'].max()

street_number = dallas_open_data_pd['Street Number'].astype('str')
street_name = dallas_open_data_pd['Street Name']
street_unit = dallas_open_data_pd['Street Unit'].fillna('').astype('str')
zip_code = dallas_open_data_pd['Zip Code']
street_address = dallas_open_data_pd['Street Address']

# Use the inspection date's month and year for inspection_month for consistency
inspection_month = dallas_open_data_pd['Inspection Date'].astype('datetime64[ns]').dt.strftime('%m/%y')

inspection_year = dallas_open_data_pd['Inspection Year']

# Check record count.  Should be 42,070
len(restaurant_name2)
```

```python
#Save data tp dataframe
clean_dallas_open_data_pd = pd.DataFrame({
    'restaurant_name0':restaurant_name0,
    #'restaurant_name1':restaurant_name1,
    'restaurant_name2':restaurant_name2,
    'restaurant_name3':restaurant_name3,
    #'restaurant_name4':restaurant_name4,
    'inspection_type':inspection_type,
    'inspection_date':inspection_date,
    'inspection_score':inspection_score,
    #'max_inspection_score':max_inspection_score,
    'street_number':street_number,
    'street_name':street_name,
    'street_unit':street_unit,
    'zip_code':zip_code,
    'street_address':street_address,
    'inspection_month':inspection_month,
    'inspection_year':inspection_year
})

# Display dataframe
clean_dallas_open_data_pd
```

```python
clean_dallas_open_data_max_score = clean_dallas_open_data_pd.groupby(['restaurant_name2'])['inspection_score'].max()
clean_dallas_open_data_max_score
```

```python
# Trying to extract unique locations.  KEEPS GIVING ME DUPLICATES!!! D'oh
#Save data tp dataframe
unique_dallas_open_data_pd = pd.DataFrame({
    'restaurant_name0':restaurant_name0,
    #'restaurant_name1':restaurant_name1,
    'restaurant_name2':restaurant_name2,
    'restaurant_name3':restaurant_name3,
    #'restaurant_name4':restaurant_name4,
    'inspection_type':inspection_type,
    'inspection_date':inspection_date,
    'inspection_score':inspection_score,
    'street_number':street_number,
    'street_name':street_name,
    'street_unit':street_unit,
    'zip_code':zip_code,
    'street_address':street_address
}).drop_duplicates(subset=['street_address'])
#.replace(r'\\n',' ', regex=True).drop_duplicates()

# Display dataframe
unique_dallas_open_data_pd#.info()
```

```python
# Output File (CSV)
unique_dallas_open_data_output = "./dallas_open_data/Unique_bus_dal_open_data_2019.csv"

# write dataframe to the csv
unique_dallas_open_data_pd.to_csv(unique_dallas_open_data_output, sep=',', encoding='utf-8', index=False)
```
